# Data Visualization
When it comes to data visualization in programming, there are numerous libraries and tools available, especially in languages like Python. Here's a glance at some popular ones:
1. **Matplotlib**: A foundational plotting library for Python,
Matplotlib offers a wide array of plots and customization options. It's often used as a base for other visualization libraries.
2. **Seaborn**: Built on top of Matplotlib, Seaborn simplifies many common visualization tasks and introduces several complex plot types. Its integration with Pandas makes it highly effective for statistical visualizations.
3. **Plotly**: Known for interactive plots, Plotly provides a dynamic user experience, allowing readers to explore data in real time.
4. **ggplot2 (for R)**: Inspired by "The Grammar of Graphics," ggplot2 is a widely-used library in R for constructing complex plots incrementally.
5. **Tableau, Power BI**: These are examples of commercial
tools used for creating visually rich dashboards and reports, often without the need for coding.

# **Scatterplot**
A scatter plot is a plot or mathematical diagram using Cartesian coordinates to display values for typically two variables for a set of data. Here each value in the data set is represented by a dot. It is used for understanding the relationship between the 2 variables.

Creating a scatterplot with matplotlib is simple, we just need to follow a simple syntax. For this plot type we use the Iris dataset.

```
plt.scatter(x, y, color='optional_color', marker='optional_marker')
plt.title('Scatterplot Title')
plt.xlabel('X-axis Label')
plt.ylabel('Y-axis Label')
plt.show()

```
Noter marker takes * . s ^ o symbols


````python
# Basic scatter plot using matplotlib
plt.figure(figsize=(10,6))
plt.scatter(df["Score"], df["Overs"], color="red", marker="*")
plt.title("ScatterPlot of runs scored and wickets lost using matplotlib")
plt.xlabel("Runs Scored")
plt.ylabel("Wickets Lost")
plt.grid()
plt.show()
````

````python
# Detailed Scatterplot using Matplotlib
# Defining colors for each species
colors = {'lost': 'red', 'n/r': 'black', 'won': 'blue', 'tied': 'green'}

markers = {'lost': 'o', 'n/r': 's', 'won': '^', 'tied': 'P'}

plt.figure(figsize=(10, 6))

for results in df['Result'].unique():
    subset = df[df['Result'] == results]
    # print(subset)
    plt.scatter(subset['Score'], subset['Overs'],
                color=colors[results], label=results, edgecolor='Yellow', marker=markers[results])


plt.title('Runs Scored vs Overs Played (Matplotlib)')
plt.xlabel('Runs Scored')
plt.ylabel('Overs Played')
plt.legend(title='Result')
plt.grid()
plt.show()
````

````python
# Scatter Plot using Seaborn
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Score', y='Overs', hue='Result', style='Result', palette='flare')
plt.title('Runs Scored vs Overs Played (Seaborn)')
plt.show()
````

The scatterplot visualizes the innings score on X axis and overs played on the Y axis. The plot shows us that Nepal has won generally when they have played fewer overs. This suggests that they have mostly won chasing a target.

# **Bar plot**
A bar chart or bar graph is a chart or graph that presents categorical data with rectangular bars with heights or lengths proportional to the values that they represent. The bars can be plotted vertically or horizontally.

For the bar plot, We visualize the species on the X axis and counts on the Y axis.

````python
# Bar Plot using Matplotlib
results = df['Result'].value_counts()
df_results = results.reset_index();

print(results)
print(df_results)

plt.figure(figsize=(10, 6))
plt.bar(results.index, results.values, color='purple')

plt.title('Results of Matches (Matplotlib)')
plt.xlabel('Results')
plt.ylabel('Count')
plt.show()
````

````python
# Barplot using Seaborn
plt.figure(figsize=(10, 6))
sns.barplot(data=df_results, x='Result', y='count', palette='cividis', hue='Result')
plt.title('Results of Matches (Seaborn)')
plt.xlabel('Results')
plt.ylabel('Count')
plt.show()
````

The bar plot shows the outcomes of matches. From the bar plot it is clear that team Nepal has won the same number of matches as they have lost. Also a few matches have had no result likely due to rain and some tied.

````python
# Plotting bar graph for results and average runs scored using matplotlib
plt.figure(figsize=(10, 6))
plt.bar(df['Result'].unique(), df.groupby('Result')['Score'].mean(), color='purple')
plt.title('Average Runs Scored by Results (Matplotlib)')
plt.xlabel('Results')
plt.ylabel('Average Runs Scored')
plt.show()
````

````python
# Plotting bar graphs for results and average runs scored using seaborn
plt.figure(figsize=(10, 6))
sns.barplot(data=df, x='Result', y='Score', palette='cividis', hue='Result')
plt.title('Average Runs Scored by Results (Seaborn)')
plt.xlabel('Results')
plt.ylabel('Average Runs Scored')
plt.show()
````

The bar plot shows the average runs scored based on result. This shows that Nepal have scored most amount of runs in winning cause which is obvious as more runs mean higher chances of winning. The average score when tied is very high and this is likely due to the amount of tied matches being very small.

# Box plots or Whisker plots

Box plot was was first introduced in year 1969 by Mathematician John Tukey.Box plot give a statical summary of the features being plotted.Top line represent the max value,top edge of box is third Quartile, middle edge represents the median,bottom edge represents the first quartile value.The bottom most line respresent the minimum value of the feature.The height of the box is called as Interquartile range.The black dots on the plot represent the outlier values in the data.

````python
# Boxplot using matplotlib
plt.figure(figsize=(10,6))
df.boxplot(column='RPO', by='Wickets Lost', grid=False)
plt.title("Box plot of RPO by Wickets Lost using matplotlib")
plt.suptitle('')
plt.xlabel("Wickets Lost")
plt.ylabel("RPO")
plt.show()
````

````python
# Box plot using seaborn
plt.figure(figsize=(10,6))
sns.boxplot(data=df, x='Wickets Lost', y='RPO', palette='cividis', hue="Wickets Lost")
plt.title("Box plot of RPO by Wickets Lost using seaborn")
plt.show()

````

The box plot shows a clear trend: as wickets lost increase, the RPO (Runs Per Over) tends to decrease. At 1-3 wickets, the RPO is higher and more spread out, particularly at 1 wicket where scores range from 4 to over 12 RPO. This suggests aggressive batting early in the innings. At 4-7 wickets, the RPO values stabilize, showing narrower spreads and medians between 4.5 to 6.5, indicating consistent scoring during the middle phase.

For 8-10 wickets, the RPO declines significantly, with medians falling to 4 or lower, reflecting slower scoring as the lower-order batsmen take over. However, a few outliers at 10 wickets hint at late aggressive hitting attempts. Overall, the plot highlights a typical batting pattern where losing more wickets reduces the scoring rate, with occasional bursts of higher RPO.

# **Heatmap**

A heatmap is a two-dimensional graphical representation of data where the individual values that are contained in a matrix are represented as colours. The Seaborn package allows the creation of annotated heatmaps which can be tweaked using Matplotlib tools as per the creator's requirement.

To create a heatmap of the Iris dataset, we typically focus on visualizing the correlation matrix, which shows the relationships between numerical features. A heatmap is a great way to visualize these correlations.

Computes the Pearson correlation matrix, showing the pairwise correlation coefficients between numerical features. Values range from:

+1: Perfect positive correlation.

0: No correlation.

-1: Perfect negative correlation.

````python
# Heatmap using matplotlib
correlation = df[['Score', 'Wickets Lost', 'Overs', 'RPO']].corr()

plt.figure(figsize=(10,6))
plt.imshow(correlation, cmap='coolwarm', interpolation='nearest')
plt.colorbar(label="Correlation Coefficient")
plt.title("Correlation Heatmap using matplotlib")
plt.xticks(range(len(correlation.columns)), correlation.columns, rotation=45)
plt.yticks(range(len(correlation.columns)), correlation.columns)
plt.show()
````

````python
# Seaborn correlation map
plt.figure(figsize=(10,6))
sns.heatmap(correlation, annot=True, cmap='coolwarm', fmt=".2f", square=True)
plt.title("Correlation Heatmap using seaborn")
plt.show()
````

- Strong correlation between score and overs. i.e. as overs increase, score increases.
- Score has little to no correlation to RPO and Wickets lost. They are not linearly related.
- RPO is negatively correlated to Wickets Lost and Overs. If one feature increases, the other decreases.


Let's check the correlation between the runs scored and target given. For this we need to drop all the rows of 1st innings i.e. drop all targets with NaN value.

# **Violin Plot**
Violin plot is a hybrid of a box plot and a kernel density plot, which shows peaks in the data. It shows the distribution of quantitative data across several levels of one (or more) categorical variables such that those distributions can be compared. Unlike a box plot, in which all of the plot components correspond to actual data points, the violin plot features a kernel density estimation of the underlying distribution.

````python
# violin plot of score distributino by innings using matplotlib
innings = sorted(df['Inns'].unique())
data = [df[df['Inns'] == i]['Score'] for i in innings]

plt.figure(figsize=(10,6))
plt.violinplot(data, showmeans=True, showmedians=True)
plt.xticks(ticks=range(1, len(innings) + 1), labels=[f'Innings {i}' for i in innings])
plt.title("Violin plot of Score distribution by Innings using matplotlib")
plt.xlabel("Innings")
plt.ylabel("Score")
plt.show()
````

The violin plot visualizes the distribution of scores across the two innings. The width of each violin indicates the density of scores at different values. For first innings, the scores are more spread out, ranging from a low near 20 to above 300, with a median score around 220. Conversely, second innings shows a slightly lower range, with scores concentrated between 50 and 290, and a median score closer to 180. Overall, first innings displays a higher variability in scores compared to second innings, suggesting a broader range of performance outcomes.

````python
# Violin plot of RPO distribution by result using seaborn
plt.figure(figsize=(10,6))
sns.violinplot(data=df, x='Result', y='RPO', palette='cividis', hue='Result')
plt.title("Violin plot of RPO distribution by Result using seaborn")
plt.xlabel("Result")
plt.ylabel("RPO")
plt.show()
````

The violin plot shows the distribution of RPO (Runs Per Over) across different match results: *lost*, *won*, *tied*, and *n/r (no result)*. 

- For lost matches, the RPO distribution is narrow, mostly between 3 and 6, with a concentration around 4.
- In won matches, the RPO has a wider spread, ranging from below 3 to beyond 12, with a median around 5. This suggests higher variability in performance, including instances of very high RPO.
- Tied matches exhibit a single RPO value near 5.5, indicating no variability in the data for this result.
- For n/r matches, the RPO is tightly concentrated between 4 and 5, showing low variability.

Overall, won matches display the widest range of RPO values, indicating that higher scoring rates (with greater spread) often correlate with wins, whereas lost matches tend to have a lower and more consistent RPO.

# **Histogram**
Histograms are used to represent the frequency distribution of continuous variables. The width of the histogram represents interval and the length represents frequency. To create a histogram you need to create bins of the interval which are not overlapping. Histogram allows the inspection of data for its underlying distribution, outliers, skewness.


````python
# Histogram of score distribution using matplotlib
plt.figure(figsize=(10,6))
plt.hist(df['Score'], bins=10)
plt.title('Histogram of Score (Matplotlib)')
plt.xlabel('Score')
plt.ylabel('Frequency')
plt.show()
````

````python
# Histogram of score distribution using seaborn
plt.figure(figsize=(10,6))
sns.histplot(data=df, x="Score", bins=10, kde=True, alpha=0.5)
plt.title("Histogram of Score (Seaborn)")
plt.xlabel("Score")
plt.ylabel("Frequency")
plt.show()
````

A left-skewed histogram suggests that the scores data has few low score values pulling the distribution to the left, while most values are concentrated on the larger side.

````python
# Histogram of Score distribution by Result using matplotlib
plt.figure(figsize=(10,6))

for result in df['Result'].unique():
    score = df[df['Result'] == result]
    plt.hist(score['Score'], bins=10, alpha=0.5, label=result)

plt.title('Histogram of Score by Result (Matplotlib)')
plt.xlabel('Score')
plt.ylabel('Frequency')
plt.legend(title='Result')
plt.show()
````

````python
# Histogram of Score distribution by Result using seaborn
plt.figure(figsize=(10,6))
sns.histplot(data=df, x='Score', bins=10, kde=True, hue='Result', palette='deep', alpha=0.5)
````

The histogram shows the distribution of scores in a dataset, grouped by game outcomes. Scores for won games are generally higher, while lost games tend to have lower scores. Tied games show a wider score distribution. The data suggests a relationship between score and game outcome, with higher scores often leading to wins and lower scores associated with losses.


# **Line Plots/Charts**
Line plots is a graph that is used for the representation of continuous data points on a number line. Line plots are created by first plotting data points on the Cartesian plane then joining those points with a number line. Line plots can help display data points for both single variable analysis as well as multiple variable analysis.

````python
plt.figure(figsize=(10,6))
plt.plot(df.index, df['Score'])
plt.title("Scores over time")
plt.xlabel("Match")
plt.ylabel("Score")
plt.show()
````

This chart shows that Nepali cricket team have been very inconsistent as the total score fluctuates time to time with no trends of improvement.

````python
plt.figure(figsize=(10,6))
sns.lineplot(data=df, x=df.index, y='Score', hue='Result', marker='o', palette='deep')
plt.title("Scores over time")
plt.xlabel("Match")
plt.ylabel("Score")
plt.show()
````

The line graph depicts the scores over time, with different colors representing different results (lost, won, tied, and n/r). The scores fluctuate over time, with some periods of higher scores and others of lower scores. The overall trend seems to be downward, with the scores decreasing over time.
